## Implementing a Simple Quantum Operation with Cirq
- **a)** Initialize a quantum system with 5 qubits
- **b)** Apply Hadamard operation on every qubit
- **c)** Apply CNOT operation on (0, 1), (1,2), (2,3), (3,4)
- **d)**  SWAP (0, 4)
- **e)**  Rotate X with pi/2 on any qubit
- **f)** Plot the circuit

```python
import cirq
import numpy as np

# Define five qubits.
q = cirq.LineQubit.range(5)

# Apply Hadamard operation on every qubit
op1 = [cirq.H(q[n]) for n in range(5)]

# Apply CNOT operation on (0, 1), (1,2), (2,3), (3,4)
op2 = [cirq.CNOT(q[0], q[1]), cirq.CNOT(q[1], q[2]), cirq.CNOT(q[2], q[3]), cirq.CNOT(q[3], q[4])]

#SWAP (0, 4)
op3 = [cirq.SWAP(q[0], q[4])]

#Rotate X with pi/2 on any qubit
op4 = [cirq.Rx(rads=0.5 * np.pi)(q[0])]

ops = op1 + op2 + op3 + op4
# Create a circuit from the list of operations.
circuit = cirq.Circuit(ops)
print(f"Circuit:\n{circuit}")
```

## Implement a second circuit with a framework of your choice:
- **a)** Apply a Hadmard gate to the first qubit
- **b)** rotate the second qubit by pi/3 around X
- **c)** Apply Hadamard gate to the third and fourth qubit
- **d)** Perform a swap test between the states of the first and second qubit |q1 q2> and the third and fourth qubit |q3 q4>

```python
import pennylane as qml
from pennylane import numpy as np

# Definition of the quantum device with 5 qubits (4 main + 1 ancilla)
dev = qml.device("default.qubit", wires=5)

# Definition of the circuit for the swap test
@qml.qnode(dev)
def circuit():
    #Apply a Hadmard gate to the first qubit
    qml.Hadamard(wires=0)
    # Rotate the second qubit by pi/3 around X
    qml.RX(np.pi/3, wires=1)
    # Apply Hadamard gate to the third and fourth qubit
    qml.Hadamard(wires=2)
    qml.Hadamard(wires=3)

    # Perform a swap test between the states of the first and second qubit |q1 q2> and the third and fourth qubit |q3 q4>t
    qml.Hadamard(wires=4)
    qml.CSWAP(wires=[4, 0, 2])
    qml.CSWAP(wires=[4, 1, 3])
    qml.Hadamard(wires=4)

    return qml.expval(qml.PauliZ(4))

# Displaying the circuit
print(qml.draw(circuit)())
```
